### modeling - simple manual model
manually look at our data and decide some cutoff points
for classification.

```python
iris_data['target_names']
```

```python
def single_feature_prediction(petal_length): # predict iris species given the petal length
    if petal_length <2.5:
        return 0
    elif petal_length <4.8:
        return 1
    else:
        return 2
    
```

```python
def plot_incorrect_predictions(df_predictions, x_axis_feature,y_axis_feature):

    fig, axs= plt.subplots(2,2, figsize=(10,10))
    axs = axs.flatten()
    sns.scatterplot(x=x_axis_feature,y=y_axis_feature,hue= 'predictions_labels', data=df_predictions, ax=axs[0])
    sns.scatterplot(x=x_axis_feature,y=y_axis_feature,hue= 'target_names', data=df_predictions, ax=axs[1])
    sns.scatterplot(x=x_axis_feature,y=y_axis_feature,hue= 'correctly_predictions', data=df_predictions, ax=axs[2])
    axs[3].set_visible(False)
    plt.show()
```

```python
plot_incorrect_predictions(df_predictions, 'petal length (cm)', 'petal width (cm)')
```
