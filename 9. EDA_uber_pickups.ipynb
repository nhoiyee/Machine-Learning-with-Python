```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv('C:/Users/Hoi Yee/Downloads/uber_pickups.csv')
```

```python
print(df.head())
```

```python
df.dropna(inplace=True)
```

```python
#convert pickup datetime to a usable format
df['pickup_time']= pd.to_datetime(df['pickup_time'])
```

```python
#extract the hour from the pickup time for analysis
df['hour']= df['pickup_time'].dt.hour
```

```python
#create a histogram to visualize the distribution of pickup by hour
sns.histplot(df['hour'], bins=24, kde=False)
plt.title('distrution of uber pickups by hour')
plt.xlabel('hour of the day')
plt.ylabel('number of pickups')
plt.show()
```

```python
#create a scatterplot to visualize the relationship between pickup time and
plt.scatter(df['hour'], df['number_of_rides'])
plt.title('scatter plot of hour vs number of uber rides')
plt.xlabel('hour of the day')
plt.ylabel('number of rides')
plt.show()
```
